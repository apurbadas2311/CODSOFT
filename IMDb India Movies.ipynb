```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Duration'], y=df['Rating'], alpha=0.6)
plt.title('Relationship Between Movie Length and Rating')
plt.xlabel('Movie Length (minutes)')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```
